```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", color_codes=True)

plt.title('Calls by Precinct', fontsize=20)
plt.xticks(rotation=90)

sns.countplot(x="Precinct", color="b", data=df);


```

```python
plt.figure(figsize=(9, 8))
sns.distplot(df['Priority'], color='g', bins=100, hist_kws={'alpha': 0.4});
```
