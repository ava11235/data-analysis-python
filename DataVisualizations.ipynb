```python
Bar Charts are used to visualize categorical variables
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('datasets/cars.csv')
```

```python
df.head()
```

```python
df.size
```

```python
sns.set(style="whitegrid", color_codes=True)

plt.title('Car Makes ', fontsize=20)

# plot a vertical bar chart with rotated x axis labels
chart = sns.countplot(x="Make", data=df);
chart.set_xticklabels(chart.get_xticklabels(), rotation=90)

#show the cart
plt.show()
```

Pie Charts are used to visualize the relative frequency of categories shown

```python
df.Make.value_counts().plot(kind='pie')
plt.axis('equal')
plt.title('Cars Makes', fontsize=20)
```

Scatter Plots are used to visualize the relationship between two variables 

```python
plt.title('City and Highway Kilowatt-hour/100', fontsize=20)

sns.regplot(df["HWY (kWh/100 km)"], df['CITY (kWh/100 km)'], ci=None, fit_reg=False);
plt.xticks(rotation=45)

plt.show()
```

Line Charts are used to visualize trends in data. Since trends happen in time, the x axis is often a time metric. 

```python
df = pd.read_csv('datasets/tesla-stock-price.csv')
df.head()

```

```python

plt.title('Tesla Stock Price 2015/10/15 - 2018/10/15', fontsize = 12)

plt.xlabel('Date')
plt.ylabel('Stock high')

plt.plot(df["date"], df["high"])
ax = plt.gca()
ax.axes.xaxis.set_visible(False)

plt.show()
```
