Pandas basics, part 2: Reshaping data

Pandas is a Python library for working with DataFrames 
Use panda's pivot and melt methods to convert dataframes from long to wide form (with pivot) and from wide to long form (with melt)

```python
import pandas as pd

# Create a DataFrame

df = pd.DataFrame({
          'Course':{0:'IT 111', 1:'IT 111', 2:'IT 111', 3:'IT 111', 4:'IT 111', 5:'IT 111'},
                  
           'Student': {0: "Abe", 1: "John", 2: "Ira", 3: "Josie", 4: "Dana", 5: "Jesse"},
            
          'Test Score': {0: 75, 1: 88, 2: 100, 3: 100, 4: 95, 5: 89}}, 
                
           columns=['Course', 'Student','Test Score'])
```

```python
df.head()
```

```python
pd.pivot(df,  index = 'Course', columns = "Student", values="Test Score")
```

```python
df1 = pd.DataFrame({'Student': {0: "Abe", 1: "John", 2: "Ira", 3: "Josie", 4: "Dana", 5: "Jesse"},
                   'Midterm Score': {0: 89, 1: 100, 2: 72, 3: 95, 4: 100, 5: 88}, 
                'Final Score': {0: 100, 1: 95, 2: 82, 3:75, 4:99, 5:85}}, 
                  columns=['Student','Midterm Score','Final Score'])
```

```python
df1.head()
```

```python
pd.melt(df1, id_vars = "Student", var_name = "Test", value_vars = ["Midterm Score", "Final Score"], value_name = "Scores")
```
